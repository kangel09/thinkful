###Exercise
For two of the following distributions:

* Normal
* Binomial
* Gamma
* Poisson

1. Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.

2. Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.


```
# this code allows the display of output from multiple calls without requiring print() each time
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
sns.set()
```

####Normal Distribution
1. Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.


```
means = []
for n in range(1, 1001):
  rand_normal = np.random.normal(0, 3, n)
  means.append(rand_normal.mean())
sns.set_style('white')
normal_stats = pd.DataFrame({'means' : means})
sns.lineplot(range(1, 1001), normal_stats['means'])
plt.axhline(normal_stats['means'].mean(), color = 'r', linestyle = 'solid', linewidth = 2)
sns.despine()
plt.title('Sample Means From the Normal Distribution\n as a Function of Sample Size')
plt.xlabel('Sample Size')
plt.ylabel('Means');
```

2. Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.



```
size = 10000
mean = 10
sd = 15
means = []
for n in range(1000):
  rand_sample = np.random.normal(mean, sd, size)
  means.append(rand_sample.mean())

sample_distribution = pd.DataFrame({'means' : means})
sns.distplot(sample_distribution['means'])
plt.title('The Sampling Distribution of 1000 Samples\n From the Normal Distribution')
plt.xlabel('Means')
plt.axvline(sample_distribution['means'].mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(sample_distribution['means'].mean() - sample_distribution['means'].std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(sample_distribution['means'].mean() + sample_distribution['means'].std(), color = 'b', linestyle = 'dashed', linewidth = 2)
sns.despine();
```

####Binomial Distribution

1. Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.







```
size = 1000
p = 0.6
n = 100
means = []
for size in range(1, 1001):
  rand_binomial = np.random.binomial(n, p, size)
  means.append(rand_binomial.mean())

sns.set_style('white')
binomial_stats = pd.DataFrame({'means' : means})
sns.lineplot(range(1, 1001), binomial_stats['means'])
plt.axhline(binomial_stats['means'].mean(), color = 'r', linestyle = 'solid', linewidth = 2)
sns.despine()
plt.title('Sample Means From the Binomial Distribution\n as a Function of Sample Size')
plt.xlabel('Sample Size')
plt.ylabel('Means');
```

2. Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

```
n_samples = 1000
p = 0.6
trials = 1000
rand_sample = np.random.binomial(trials, p, n_samples)

sample_distribution = pd.DataFrame({'means' : rand_sample})
sns.distplot(sample_distribution['means'])
plt.title('The Sampling Distribution of 1000 Samples\n From the Binomial Distribution: p = {}, n = {}'.format(p, n))
plt.xlabel('Means')
plt.axvline(sample_distribution['means'].mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(sample_distribution['means'].mean() - sample_distribution['means'].std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(sample_distribution['means'].mean() + sample_distribution['means'].std(), color = 'b', linestyle = 'dashed', linewidth = 2)
sns.despine();
```

####Gamma Distribution

1. Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.







```
shape = 3
scale = 5
means = []
for size in range(1, 1001):
  rand_gamma = np.random.gamma(shape, scale, size)
  means.append(rand_gamma.mean())

sns.set_style('white')
gamma_stats = pd.DataFrame({'means' : means})
sns.lineplot(range(1, 1001), gamma_stats['means'])
plt.axhline(gamma_stats['means'].mean(), color = 'r', linestyle = 'solid', linewidth = 2)
sns.despine()
plt.title('Sample Means From the Gamma Distribution\n as a Function of Sample Size')
plt.xlabel('Sample Size')
plt.ylabel('Means');
```

2. Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

```
size = 1000
n_samples = 1000
shape = 3
scale = 5
sample_means = []
for n in range(n_samples):
  rand_sample = np.random.gamma(shape, scale, size)
  sample_means.append(rand_sample.mean())

sample_distribution = pd.DataFrame({'means' : sample_means})
sns.distplot(sample_distribution['means'])
plt.title('The Sampling Distribution of 1000 Samples\n From the Gamma Distribution: alpha = {}, beta = {}'.format(shape, scale))
plt.xlabel('Means')
plt.axvline(sample_distribution['means'].mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(sample_distribution['means'].mean() - sample_distribution['means'].std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(sample_distribution['means'].mean() + sample_distribution['means'].std(), color = 'b', linestyle = 'dashed', linewidth = 2)
sns.despine();
```

####Poisson Distribution

1. Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.







```
lam = 3
means = []
for size in range(1, 1001):
  rand_poisson = np.random.poisson(lam, size)
  means.append(rand_poisson.mean())

sns.set_style('white')
poisson_stats = pd.DataFrame({'means' : means})
sns.lineplot(range(1, 1001), poisson_stats['means'])
plt.axhline(poisson_stats['means'].mean(), color = 'r', linestyle = 'solid', linewidth = 2)
sns.despine()
plt.title('Sample Means From the Poisson Distribution\n as a Function of Sample Size')
plt.xlabel('Sample Size')
plt.ylabel('Means');
```

2. Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

```
size = 1000
n_samples = 1000
lam = 3
sample_means = []
for n in range(n_samples):
  rand_sample = np.random.poisson(lam, size)
  sample_means.append(rand_sample.mean())

sample_distribution = pd.DataFrame({'means' : sample_means})
sns.distplot(sample_distribution['means'])
plt.title('The Sampling Distribution of 1000 Samples\n From the Lambda Distribution: lambda = {}'.format(lam))
plt.xlabel('Means')
plt.axvline(sample_distribution['means'].mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(sample_distribution['means'].mean() - sample_distribution['means'].std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(sample_distribution['means'].mean() + sample_distribution['means'].std(), color = 'b', linestyle = 'dashed', linewidth = 2)
sns.despine();
```
