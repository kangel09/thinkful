```
# this code allows the display of output from multiple calls without requiring print() each time
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
sns.set()
%matplotlib inline
```

####**The Normal Distribution**

```
normal_var = np.random.normal(5, 3, 100)
sns.set_style("white")
sns.distplot(normal_var)
plt.axvline(normal_var.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(normal_var.mean() + normal_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(normal_var.mean() - normal_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2) 
plt.title('The Normal Distribution, mean = 5, sd = 3')
sns.despine();
```

####**The Binomial Distribution**

```
binomial_var = np.random.binomial(5, 0.7, 100)
sns.distplot(binomial_var, kde=False)
plt.axvline(binomial_var.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(binomial_var.mean() + binomial_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(binomial_var.mean() - binomial_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2);
plt.title('The Binomial Distribution, n = 5, p = 0.7');
```

####**The Poisson Distribution**

```
poisson_var = np.random.poisson(lam = 3, size = 100)
sns.distplot(poisson_var, kde=False)
plt.axvline(poisson_var.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(poisson_var.mean() + poisson_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(poisson_var.mean() - poisson_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2);
plt.title('The Poisson Distribution, lambda = 3');
```

####**The Exponential Distribution**

```
exponential_var = np.random.exponential(1/3, 100)
sns.distplot(exponential_var, kde=False)
plt.axvline(exponential_var.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(exponential_var.mean() + exponential_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(exponential_var.mean() - exponential_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2);
plt.title('The Exponential Distribution, lambda = 1/3');
```

####**The Gamma Distribution**

```
gamma_var = np.random.gamma(3, 5, 100)
sns.distplot(gamma_var, kde=False)
plt.axvline(gamma_var.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(gamma_var.mean() + gamma_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(gamma_var.mean() - gamma_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2);
plt.title('The Gamma Distribution, alpha - 3, beta = 5');
```

####**The Beta Distribution**

```
beta_var = np.random.beta(0.9, 20, 100)
sns.distplot(beta_var, kde=False)
plt.axvline(beta_var.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(beta_var.mean() + beta_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(beta_var.mean() - beta_var.std(), color = 'b', linestyle = 'dashed', linewidth = 2);
plt.title('The Beta Distribution, n = 5, p = 0.7');
```

###Exercise 2
Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and a standard deviation of 1.

Add them together to create a third variable.

Graph the third variable using a histogram.

Compute the mean and standard deviation and plot them as vertical lines on the histogram.

Evaluate the descriptive statistics against the data.

```
normal_1 = np.random.normal(5, 0.5, 1000)
normal_2 = np.random.normal(10, 1, 1000)
normal_3 = normal_1 + normal_2
sns.distplot(normal_3, kde = False)
plt.axvline(normal_3.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(normal_3.mean() - normal_3.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(normal_3.mean() + normal_3.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.title('The Sum of Two Normal Random Variables');
```
